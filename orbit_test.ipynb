## 1. A circular orbit

Use the virial theorem to calculate circular velocity:

$$ v^2 = \frac{GM}{R} $$

```python
## Constants for orbits (SI units)
## const.G, const.M_jup, counts.au

G  = const.G.value      # m^3 kg^-1 s^-2
MJ = const.M_jup.value  # kg
MS = const.M_sun.value  # kg
AU = const.au.value     # m
```

```python
# vel = sqrt(GM/R)
# R is in units of AU
def v_circ ( R ):
    return G * MS / (AU * R)  # km/s
```

```python
def circ_orbit ( nsteps, R ):
    period = 2.0 * np.pi * R * AU / v_circ(R)  # s
    theta  = (2.0 * np.pi / period) * np.linspace(0, period, nsteps)
    x      = AU * R * np.cos(theta)
    y      = AU * R * np.sin(theta)
    return (x, y)
```

```python
R_jup         = 5   # au
orbit_jupiter = circ_orbit( 100, R_jup )

plt.plot( orbit_jupiter[0], orbit_jupiter[1], 'b-' )
```

## 2. Circular orbits (3-D)

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
def get_3d_coords( x, y, theta):
    xnew = x * np.cos(theta)
    ynew = y * np.cos(theta)
    znew = x * np.sin(theta)
    return (xnew, ynew, znew)
```

```python
test_orbits = [1.0, 2.0, 3.0, 4.0]
rand_theta  = np.random.uniform(size=len(test_orbits))
theta       = dict(zip(test_orbits, rand_theta))

NSTEPS      = 100
COLORS      = dict(zip(test_orbits, ['k','b','r','g']))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for oo in test_orbits:
    (x, y) = circ_orbit(NSTEPS, oo)
    (xp, yp, zp) = get_3d_coords(x, y, theta[oo])
    ax.plot(xp, yp, zp, ls='-', lw=1, color=COLORS[oo])
```

## 3. Arbitrary orbit (2-D)

Calculate based on the energy

$$ E = - \frac{G M m}{2 a} $$

$$ v = \sqrt{ \frac{GM}{a} \left( \frac{2a}{r} - 1 \right)} $$

In polar coordinates:
$$ r^2 = 1 + e^2 \sin^2 \theta t $$

cartesian:
$$ x = a \cos \theta t $$
$$ y = b \sin \theta t $$

$$ e = \sqrt(a^2 - b^2) / b $$

```python
# a : semi-major axis [AU]
# r : radius [AU]
#
def v_ellipse( x0, y0, a ):
    a_au = a * AU
    r_au = (x0**2 + y0**2) * AU
    term1 = G * MS / (a_au)
    term2 = ((2 * a_au / r_au) - 1 )
    if term2 < 0:
        print "ERROR: a is too small"
        return
    return np.sqrt(term1 * term2)

def T_ellipse( a ):
    arg = 4 * np.pi**2 * (a*AU)**3 / (G * MS)
    return np.sqrt(arg) # s

def calc_theta( x, y ):
    if (x < 0) and (y > 0):
        return np.arctan(y/x) + np.pi
    if (x < 0) and (y < 0):
        return np.arctan(y/x) + np.pi
    return np.arctan(y/x)
```

```python
class Position (object):
    def __init__(self, x=0.0, y=0.0 ):
        self.x   = x
        self.y   = y
    
    @property
    def theta(self):
        return calc_theta(self.x, self.y)
    
    def iter_pos(self, accel, vel, delta_t ):
        self.x   += (vel.vx + 0.5 * accel.ax * delta_t) * delta_t
        self.y   += (vel.vy + 0.5 * accel.ay * delta_t) * delta_t

class Velocity (object):
    def __init__(self, vx=0.0, vy=0.0 ):
        self.vx   = vx
        self.vy   = vy
    def iter_vel(self, accel, delta_t ):
        self.vx   += accel.ax * delta_t
        self.vy   += accel.ay * delta_t

class Accel (object):
    def __init__(self, x, y):
        rad     = np.sqrt( (x*AU)**2 + (y*AU)**2 )
        force   = - G * MS / rad**2
        self.ax = force * np.cos( calc_theta(x, y) )
        self.ay = force * np.sin( calc_theta(x, y) )
```

```python
def move_forward( pos, vel, delta_t ):
    new_accel = Accel( pos.x, pos.y )
    pos.iter_pos( new_accel, vel, delta_t )
    vel.iter_vel( new_accel, delta_t )
    return
```

```python
xvals = [1.0]
yvals = [2.0]
A0    = 3.0
nsteps = 100

position = Position( xvals[0], yvals[0] )

print calc_theta( position.x, position.y )

v0       = v_ellipse( xvals[0], yvals[0], A0 )
velocity = Velocity( -v0 * np.cos(position.theta), v0 * np.sin(position.theta) )
time     = np.linspace(0, T_ellipse(A0), 100 )
delta_t  = T_ellipse(A0) / nsteps

print T_ellipse(A0)
```

```python
for i in range(nsteps-1):
    move_forward( position, velocity, delta_t )
    xvals.append( position.x )
    yvals.append( position.y )
```

```python
plt.plot( time, xvals, 'b-', lw=2 )
plt.plot( time, yvals, 'r-', lw=2 )
```
